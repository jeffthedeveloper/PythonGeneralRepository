```python
import pyautogui as openFilesPosition

# The hotkey method allows us to execute multiple keyboard shortcuts
# This is the same as pressing Windows + R
openFilesPosition.hotkey('win', 'r')

# Time to wait for the computer to process
openFilesPosition.sleep(2)

# Type 'Notepad'
openFilesPosition.typewrite('notepad')

# Time to wait for the computer to process
openFilesPosition.sleep(2)

# Press the Enter key
openFilesPosition.press('enter')

# Time to wait for the computer to process
openFilesPosition.sleep(3)

# Type a message in Notepad
openFilesPosition.typewrite('We opened Notepad with a robot or script')

# Time to wait for the computer to process
openFilesPosition.sleep(2)

# Get the active window
closeNotepadWindow = openFilesPosition.getActiveWindow()

# Time to wait for the computer to process
openFilesPosition.sleep(2)

# Close the active window
closeNotepadWindow.close()

# Time to wait for the computer to process
openFilesPosition.sleep(2)

# Press the TAB key
openFilesPosition.press('tab')

# Time to wait for the computer to process
openFilesPosition.sleep(2)

# Press the Enter key
openFilesPosition.press('enter')
```

```python
import pyautogui
import pyautogui as chooseOption

option = pyautogui.confirm('Click the desired button', buttons = ['Excel', 'Word', 'Notepad'])

if option == 'Excel':
    
    # The hotkey method allows us to execute multiple keyboard shortcuts
    # This is the same as pressing Windows + R
    chooseOption.hotkey('win', 'r')
    
    # Time to wait for the computer to process
    chooseOption.sleep(2)
    
    # Type 'Excel'
    chooseOption.typewrite('Excel')
    
    # Time to wait for the computer to process
    chooseOption.sleep(2)
    
    # Press the Enter key
    chooseOption.press('Enter')
    
    # Time to wait for the computer to process
    chooseOption.sleep(4)
    
    # Click the option to open a blank Excel workbook
    chooseOption.click(x=600, y=241)
    
    # Time to wait for the computer to process
    chooseOption.sleep(3)
    
    chooseOption.typewrite('I chose to open Excel')
    
elif option == 'Word':
    
    # The hotkey method allows us to execute multiple keyboard shortcuts
    # This is the same as pressing Windows + R
    chooseOption.hotkey('win', 'r')
    
    # Time to wait for the computer to process
    chooseOption.sleep(2)
    
    # Type 'winword'
    chooseOption.typewrite('winword')
    
    # Time to wait for the computer to process
    chooseOption.sleep(2)
    
    # Press the Enter key
    chooseOption.press('Enter')
    
    # Time to wait for the computer to process
    chooseOption.sleep(6)
    
    # Click the option to open a blank Word document
    chooseOption.click(x=447, y=334)
    
    # Time to wait for the computer to process
    chooseOption.sleep(5)
    
    chooseOption.typewrite('I chose to open Word')
    
elif option == 'Notepad':
    
    # The hotkey method allows us to execute multiple keyboard shortcuts
    # This is the same as pressing Windows + R
    chooseOption.hotkey('win', 'r')
    
    # Time to wait for the computer to process
    chooseOption.sleep(2)
    
    # Type 'Notepad'
    chooseOption.typewrite('Notepad')
    
    # Time to wait for the computer to process
    chooseOption.sleep(2)
    
    # Press the Enter key
    chooseOption.press('Enter')
    
    # Time to wait for the computer to process
    chooseOption.sleep(4)
    
    chooseOption.typewrite('I chose to open Notepad')
```
